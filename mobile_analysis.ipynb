## data cleaning

ok so the RAM column has values like '8 GB' and battery has '5000 mAh, Li-Po Battery' which are strings not numbers. i need to extract just the number part so i can actually use them for analysis.

also i wont delete any rows, just fix the values

```python
mob = mobile.copy()
```

```python
# RAM - extract the number from strings like '8 GB'
mob['RAM_GB'] = mob['RAM'].str.extract(r'(\d+)').astype(float)
mob[['RAM', 'RAM_GB']].head(6)
```

```python
# storage - same thing
mob['Storage_GB'] = mob['Storage'].str.extract(r'(\d+)').astype(float)
mob[['Storage', 'Storage_GB']].head(6)
```

```python
# battery - '5000 mAh, Li-Po' -> 5000
mob['Battery_mAh'] = mob['Battery'].str.extract(r'(\d+)').astype(float)
mob[['Battery', 'Battery_mAh']].head(6)
```

```python
# fill missing text with 'Unknown' instead of NaN
mob['Type']  = mob['Type'].fillna('Unknown')
mob['Technical.OS'] = mob['Technical.OS'].fillna('Unknown')
mob['Storage'] = mob['Storage'].fillna('Unknown')
mob['RAM'] = mob['RAM'].fillna('Unknown')
mob['Battery'] = mob['Battery'].fillna('Unknown')
```

```python
# fill missing numbers with median 
mob['RAM_GB'] = mob['RAM_GB'].fillna(mob['RAM_GB'].median())
mob['Storage_GB']  = mob['Storage_GB'].fillna(mob['Storage_GB'].median())
mob['Battery_mAh'] = mob['Battery_mAh'].fillna(mob['Battery_mAh'].median())
```

```python
# adding a price category column - this will be useful for graphs later
mob['Price_Category'] = 'Flagship (>80k)'
mob.loc[mob['Price'] <= 80000, 'Price_Category'] = 'Premium (40-80k)'
mob.loc[mob['Price'] <= 40000, 'Price_Category'] = 'Mid-range (20-40k)'
mob.loc[mob['Price'] <= 20000, 'Price_Category'] = 'Economy (10-20k)'
mob.loc[mob['Price'] <= 10000, 'Price_Category'] = 'Budget (<10k)'
```

```python
# verify no nulls left
mob[['RAM_GB', 'Storage_GB', 'Battery_mAh', 'Price_Category']].isnull().sum()
```

## matplotlib graphs

```python
top10 = mob['Brand'].value_counts().head(10)
plt.figure(figsize=(10, 5))
plt.bar(top10.index, top10.values, color='darkgreen')
plt.title('Top 10 Brands by Number of Models')
plt.xlabel('Brand')
plt.ylabel('Number of Models')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

avg_price = mob.groupby('Brand')['Price'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 5))
plt.bar(avg_price.index, avg_price.values, color='steelblue')
plt.title('Top 10 Brands by Average Price')
plt.xlabel('Brand')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

price_counts = mob['Price_Category'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(price_counts.values, labels=price_counts.index, autopct='%1.1f%%', colors=['skyblue', 'orange', 'lightgreen', 'salmon'])
plt.title('Phone Distribution by Price Category')
plt.tight_layout()
plt.show()
```

```python
# price distribution
prices = mob[mob['Price'] <= 100000]['Price']

plt.figure(figsize=(10, 5))
plt.hist(prices, bins=50, color='coral', edgecolor='white')
plt.title('Price Distribution (phones upto 1 lakh)')
plt.xlabel('Price')
plt.ylabel('Count')
plt.show()
```

```python
# ram sizes bar chart
ram = mob['RAM_GB'].value_counts().sort_index()
ram = ram[ram >= 30]

plt.figure(figsize=(10, 5))
plt.bar(ram.index.astype(int).astype(str), ram.values, color='mediumpurple')
plt.title('Most Common RAM Sizes')
plt.xlabel('RAM (GB)')
plt.ylabel('Number of Phones')
plt.tight_layout()
plt.show()
```

## seaborn graphs

seaborn makes nicer looking graphs with less code

```python
top10_brands = mob['Brand'].value_counts().head(10).index
top10_df = mob[mob['Brand'].isin(top10_brands)]

plt.figure(figsize=(11, 5))
sns.countplot(data=top10_df, x='Brand',hue ="Brand",order=top10_brands, palette='tab10',legend=False)
plt.title('Top 10 Brands')
plt.xlabel('Brand')
plt.ylabel('Count')
plt.show()
```

```python
# box plot for price in categories
under1L = mob[mob['Price'] <= 100000]
cat_order = ['Budget (<10k)', 'Economy (10-20k)', 'Mid-range (20-40k)', 'Premium (40-80k)', 'Flagship (>80k)']

plt.figure(figsize=(12, 6))
sns.boxplot(data=under1L, x='Price_Category', y='Price',hue='Price_Category', order=cat_order, palette='Set2', legend=False)
plt.title('Price Range per Category')
plt.xlabel('Category')
plt.ylabel('Price')
plt.show()
```

```python
# scatter 
sample = mob[(mob['Price'] <= 100000) & (mob['RAM_GB'] <= 16)].sample(1500, random_state=42)

plt.figure(figsize=(9, 6))
sns.scatterplot(data=sample, x='RAM_GB', y='Price', hue='Price_Category', alpha=0.5, s=30)
plt.title('RAM vs Price')
plt.xlabel('RAM (GB)')
plt.ylabel('Price')
plt.legend(title='Category', bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
```

```python
# heatmap - correlation between all numeric columns
corr = mob[['Price', 'RAM_GB', 'Storage_GB', 'Battery_mAh']].corr()

plt.figure(figsize=(7, 5))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```
